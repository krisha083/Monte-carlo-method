```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(num_samples):
    # Generate random points
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)

    # Compute distance from origin
    distance = np.sqrt(x**2 + y**2)

    # Count points inside the quarter circle
    inside_circle = np.sum(distance <= 1)

    # Estimate pi
    pi_estimate = (inside_circle / num_samples) * 4

    return pi_estimate

# Number of samples
num_samples = 10000

# Run the Monte Carlo simulation
pi_estimate = monte_carlo_pi(num_samples)

print(f"Estimated value of Ï€ with {num_samples} samples: {pi_estimate}")

# Optional: Plotting the points
x = np.random.uniform(-1, 1, num_samples)
y = np.random.uniform(-1, 1, num_samples)
plt.figure(figsize=(6, 6))
plt.scatter(x, y, c=np.sqrt(x**2 + y**2) <= 1, cmap='bwr', s=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Monte Carlo Simulation for Ï€')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
